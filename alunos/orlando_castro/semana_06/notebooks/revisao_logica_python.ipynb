## 🔷 Bloco 3 — Operadores

### Operadores Aritméticos (contas)

| Operador | Significado | Exemplo |
|----------|-------------|---------|
| `+` | Soma | `10 + 5 = 15` |
| `-` | Subtração | `10 - 3 = 7` |
| `*` | Multiplicação | `4 * 3 = 12` |
| `/` | Divisão | `10 / 4 = 2.5` |
| `//` | Divisão inteira | `10 // 4 = 2` |
| `%` | Resto da divisão | `10 % 3 = 1` |
| `**` | Potência | `2 ** 8 = 256` |

````python
# =============================================
# 🏭 CENÁRIO: Calculando indicadores de produção
# =============================================

pecas_produzidas = 1500
pecas_com_defeito = 37
horas_trabalhadas = 8
numero_operadores = 6
custo_por_peca = 12.50  # reais

# Peças boas
pecas_boas = pecas_produzidas - pecas_com_defeito
print(f"Peças boas: {pecas_boas}")

# Taxa de defeito (em %)
taxa_defeito = (pecas_com_defeito / pecas_produzidas) * 100
print(f"Taxa de defeito: {taxa_defeito:.2f}%")

# Produtividade por operador
produtividade = pecas_produzidas / numero_operadores
print(f"Peças por operador: {produtividade:.1f}")

# Produção por hora
pecas_por_hora = pecas_produzidas / horas_trabalhadas
print(f"Peças por hora: {pecas_por_hora:.1f}")

# Faturamento do turno
faturamento = pecas_boas * custo_por_peca
print(f"Faturamento do turno: R$ {faturamento:,.2f}")
````

## 🔷 Bloco 6 — Estruturas de Repetição

### 6.1 — Loop `for`
Usamos `for` quando sabemos **quantas vezes** queremos repetir algo.

````python
# =============================================
# 🏭 CENÁRIO: Relatório diário de produção
# =============================================

# Lista de produção por dia da semana
producao_semanal = [1450, 1380, 1520, 1490, 1600]
dias = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"]

print("=== PRODUÇÃO SEMANAL ===")

total = 0
for i in range(len(dias)):
    print(f"{dias[i]}: {producao_semanal[i]} peças")
    total += producao_semanal[i]  # acumula o total

media = total / len(dias)
print(f"\nTotal da semana: {total} peças")
print(f"Média diária: {media:.1f} peças")
````

````python
# Usando range() — gera sequências de números
print("Contagem de 1 a 5:")
for numero in range(1, 6):
    print(numero)

print("\nTabuada do 3:")
for i in range(1, 11):
    resultado = 3 * i
    print(f"3 x {i:2} = {resultado:3}")
````

````python
# for com if dentro — combinando tudo!
producao_semanal = [1450, 1380, 1520, 1490, 1600]
dias = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"]
meta_diaria = 1500

print("=== ANÁLISE DA META ===")
dias_acima = 0
dias_abaixo = 0

for i in range(len(dias)):
    producao = producao_semanal[i]
    if producao >= meta_diaria:
        print(f"✅ {dias[i]}: {producao} — META BATIDA")
        dias_acima += 1
    else:
        diferenca = meta_diaria - producao
        print(f"❌ {dias[i]}: {producao} — faltaram {diferenca} peças")
        dias_abaixo += 1

print(f"\nDias acima da meta: {dias_acima}")
print(f"Dias abaixo da meta: {dias_abaixo}")
````

### 6.2 — Loop `while`
Usamos `while` quando queremos repetir **enquanto uma condição for verdadeira**.

> ⚠️ **Cuidado com loops infinitos!** A condição precisa se tornar falsa em algum momento.

````python
# =============================================
# 🏭 CENÁRIO: Simulando uma esteira de produção
# =============================================

estoque_materias_primas = 100  # kg de matéria-prima
consumo_por_lote = 15          # kg consumidos por lote
lotes_produzidos = 0

print("=== SIMULAÇÃO DE PRODUÇÃO ===")
print(f"Estoque inicial: {estoque_materias_primas} kg\n")

while estoque_materias_primas >= consumo_por_lote:
    estoque_materias_primas -= consumo_por_lote
    lotes_produzidos += 1
    print(f"Lote {lotes_produzidos} produzido | Estoque restante: {estoque_materias_primas} kg")

print(f"\n⚠️  Matéria-prima insuficiente para novo lote")
print(f"Total de lotes produzidos: {lotes_produzidos}")
print(f"Sobra de material: {estoque_materias_primas} kg")
````

## 🔷 Bloco 8 — Funções

Funções são **blocos de código reutilizáveis**. Em vez de repetir o mesmo código, você cria uma função e chama ela quando precisar.

```python
def nome_da_funcao(parametro1, parametro2):
    # código aqui
    return resultado
```

````python
# =============================================
# 🏭 CENÁRIO: Funções para indicadores de qualidade
# =============================================

def calcular_taxa_defeito(total_produzido, total_defeituoso):
    """Calcula a taxa de defeito em porcentagem."""
    if total_produzido == 0:
        return 0
    taxa = (total_defeituoso / total_produzido) * 100
    return taxa


def classificar_qualidade(taxa_defeito):
    """Classifica a qualidade com base na taxa de defeito."""
    if taxa_defeito <= 1.0:
        return "EXCELENTE"
    elif taxa_defeito <= 3.0:
        return "BOM"
    elif taxa_defeito <= 5.0:
        return "ATENÇÃO"
    else:
        return "CRÍTICO"


def gerar_relatorio_turno(turno, operador, pecas, defeitos):
    """Gera um relatório completo do turno."""
    taxa = calcular_taxa_defeito(pecas, defeitos)
    classificacao = classificar_qualidade(taxa)
    pecas_boas = pecas - defeitos
    
    print(f"\n{'='*40}")
    print(f"  RELATÓRIO — TURNO {turno.upper()}")
    print(f"{'='*40}")
    print(f"  Operador      : {operador}")
    print(f"  Peças prod.   : {pecas}")
    print(f"  Peças boas    : {pecas_boas}")
    print(f"  Defeituosas   : {defeitos}")
    print(f"  Taxa defeito  : {taxa:.2f}%")
    print(f"  Classificação : {classificacao}")
    print(f"{'='*40}")


# Chamando as funções
gerar_relatorio_turno("Manhã", "Ana Silva", 520, 8)
gerar_relatorio_turno("Tarde", "Bruno Costa", 495, 28)
gerar_relatorio_turno("Noite", "Carla Souza", 485, 4)
````

### ✏️ Atividade Acompanhada 4 — Crie sua função
Crie uma função que calcule o OEE (Overall Equipment Effectiveness) — um dos indicadores mais importantes na indústria:

````python
# ✏️ ATIVIDADE ACOMPANHADA 4
# OEE = Disponibilidade × Performance × Qualidade × 100
# Cada fator vai de 0 a 1 (ex: 0.85 = 85%)

def calcular_oee(disponibilidade, performance, qualidade):
    """
    Calcula o OEE (Overall Equipment Effectiveness)
    
    Parâmetros:
        disponibilidade: tempo disponível / tempo planejado (0 a 1)
        performance: velocidade real / velocidade ideal (0 a 1)
        qualidade: peças boas / total produzido (0 a 1)
    """
    # TODO: calcule o OEE
    oee = disponibilidade * performance * qualidade * 100
    return oee


# Testando a função
oee_resultado = calcular_oee(
    disponibilidade=0.95,  # máquina disponível 90% do tempo
    performance=0.92,      # operando a 92% da velocidade ideal
    qualidade=0.98         # 98% de peças boas
)

print(f"OEE calculado: {oee_resultado:.1f}%")

# OEE de classe mundial é ≥ 85%
if oee_resultado >= 85:
    print("🏆 Equipamento de CLASSE MUNDIAL!")
else:
    print("📈 Há oportunidade de melhoria")
````

````python
# ============================================================
# 📝 ATIVIDADE INDIVIDUAL 1 — Calculadora de Salário
# ============================================================
# 
# ENUNCIADO:
# Um operador recebe R$ 18,50 por hora normal.
# Horas extras (acima de 44h semanais) valem 50% a mais.
#
# Crie um programa que:
# 1. Peça o nome do funcionário
# 2. Peça as horas trabalhadas na semana
# 3. Calcule o salário semanal (com horas extras se houver)
# 4. Exiba um comprovante formatado
#
# DICA:
# - Salário normal = horas normais x R$ 18,50
# - Horas extras = horas trabalhadas - 44 (se > 44)
# - Valor hora extra = 18,50 x 1,5
# ============================================================

# SEU CÓDIGO AQUI:

nome = input("Digite o nome do funcionário: ")
horas_trabalhadas = float(input("Digite as horas trabalhadas na semana: "))

salario_normal = 18.50 * horas_trabalhadas
horas_extras = max(0, horas_trabalhadas - 44)    # calcula as horas extras (se houver)
valor_hora_extra = 18.50 * 1.5
salario_extras = horas_extras * valor_hora_extra
salario_total = salario_normal + salario_extras

print("=" * 30)
print(f"Comprovante de pagamento para {nome}:")
print("=" * 30)
print(f"Horas extras   : {horas_extras:.2f} horas")
print(f"Salário normal : R$ {salario_normal:.2f}")
print(f"Salário extras : R$ {salario_extras:.2f}")
print(f"Salário total  : R$ {salario_total:.2f}")
````

````python
# ============================================================
# 📝 ATIVIDADE INDIVIDUAL 2 — Verificador de Lote
# ============================================================
#
# ENUNCIADO:
# Um lote de peças passa pelo controle de qualidade.
# Você tem a lista de medições de diâmetro (em mm) abaixo.
# A especificação permite entre 49.8mm e 50.2mm.
#
# Crie um programa que:
# 1. Percorra a lista de medições
# 2. Conte quantas peças estão DENTRO e FORA da especificação
# 3. Calcule o percentual de aprovação
# 4. Classifique o lote:
#    - Aprovado: aprovação >= 98%
#    - Aprovado c/ ressalvas: entre 95% e 97.9%
#    - Reprovado: < 95%
# ============================================================

medicoes = [50.1, 49.9, 50.0, 50.3, 49.7, 50.2, 50.1, 49.8,
            50.0, 50.1, 49.6, 50.2, 50.0, 49.9, 50.1, 50.4,
            49.8, 50.0, 50.1, 49.9]

especificacao_min = 49.8
especificacao_max = 50.2

# SEU CÓDIGO AQUI:
pecas_dentro = 0
pecas_fora = 0
for medida in medicoes:
    if especificacao_min <= medida <= especificacao_max:
        pecas_dentro += 1
    else:
        pecas_fora += 1
total_pecas = len(medicoes)
percentual_aprovacao = (pecas_dentro / total_pecas) * 100   
print(f"Peças dentro da especificação: {pecas_dentro}")
print(f"Peças fora da especificação: {pecas_fora}")
print(f"Percentual de aprovação: {percentual_aprovacao:.2f}%")
if percentual_aprovacao >= 98:
    classificacao = "✅Aprovado"
elif percentual_aprovacao >= 95:
    classificacao = "⚠️Aprovado c/ ressalvas"
else:
    classificacao = "❌Reprovado"
print(f"Classificação do lote: {classificacao}")

````

### 🔹 Atividade Individual 3 — Função de Conversão
**Dificuldade:** ⭐⭐⭐ (Difícil)

````python
# ============================================================
# 📝 ATIVIDADE INDIVIDUAL 3 — Funções de Conversão de Unidades
# ============================================================
#
# ENUNCIADO:
# Na indústria trabalhamos com diferentes unidades de medida.
# Crie um programa com FUNÇÕES para converter:
#
# a) Temperatura: Celsius ↔ Fahrenheit
#    F = (C × 9/5) + 32
#    C = (F - 32) × 5/9
#
# b) Pressão: bar → PSI e PSI → bar
#    1 bar = 14.504 PSI
#
# c) Velocidade: m/s → km/h e km/h → m/s
#    1 m/s = 3.6 km/h
#
# Depois crie um menu simples com while para o usuário
# escolher qual conversão fazer.
# ============================================================

# SEU CÓDIGO AQUI:
def celsius_para_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_para_celsius(f):
    return (f - 32) * 5/9

def bar_para_psi(bar):
    return bar * 14.504

def psi_para_bar(psi):
    return psi / 14.504

def ms_para_kmh(ms):
    return ms * 3.6

def kmh_para_ms(kmh):
    return kmh / 3.6

def celsius_para_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_para_celsius(f):
    return (f - 32) * 5/9

def bar_para_psi(bar):
    return bar * 14.504

def psi_para_bar(psi):
    return psi / 14.504

def ms_para_kmh(ms):
    return ms * 3.6

def kmh_para_ms(kmh):
    return kmh / 3.6

while True:  # ← APENAS UM while
    print("\nMenu de Conversão:")
    print("1. Celsius → Fahrenheit")
    print("2. Fahrenheit → Celsius")
    print("3. Bar → PSI")
    print("4. PSI → Bar")
    print("5. m/s → km/h")
    print("6. km/h → m/s")
    print("0. Sair")
    escolha = input("Escolha uma opção: ")
    if escolha == "1":
        c = float(input("Digite a temperatura em Celsius: "))
        f = celsius_para_fahrenheit(c)
        print(f"{c:.2f}°C = {f:.2f}°F")
    elif escolha == "2":
        f = float(input("Digite a temperatura em Fahrenheit: "))
        c = fahrenheit_para_celsius(f)
        print(f"{f:.2f}°F = {c:.2f}°C")
    elif escolha == "3":
        bar = float(input("Digite a pressão em bar: "))
        psi = bar_para_psi(bar)
        print(f"{bar:.2f} bar = {psi:.2f} PSI")
    elif escolha == "4":
        psi = float(input("Digite a pressão em PSI: "))
        bar = psi_para_bar(psi)
        print(f"{psi:.2f} PSI = {bar:.2f} bar")
    elif escolha == "5":
        ms = float(input("Digite a velocidade em m/s: "))
        kmh = ms_para_kmh(ms)
        print(f"{ms:.2f} m/s = {kmh:.2f} km/h")
    elif escolha == "6":
        kmh = float(input("Digite a velocidade em km/h: "))
        ms = kmh_para_ms(kmh)
        print(f"{kmh:.2f} km/h = {ms:.2f} m/s")
    elif escolha == "0":
        print("Encerrando o programa. Até mais!")
        break
    else:
        print("Opção inválida. Tente novamente.")
````

### 🔶 Atividade em Grupo 1 — Sistema de Registro de Produção
**Grupo:** 3-4 pessoas | **Dificuldade:** ⭐⭐☆

````python
# ============================================================
# 👥 ATIVIDADE EM GRUPO 1 — Sistema de Registro de Produção
# ============================================================
#
# ENUNCIADO:
# Vocês devem criar um mini-sistema de registro de produção.
# O sistema deve:
#
# 1. Usar um while para manter o programa rodando
# 2. Apresentar um menu com as opções:
#    [1] Registrar produção de um turno
#    [2] Ver todos os registros
#    [3] Ver estatísticas (total, maior, menor, média)
#    [4] Sair
#
# 3. Armazenar os registros em uma lista de dicionários
# 4. Cada registro deve conter:
#    - nome do operador
#    - turno (Manhã/Tarde/Noite)
#    - quantidade produzida
#    - quantidade com defeito
#    - taxa de defeito (calculada automaticamente)
#
# DICA: Dividam as tarefas! Uma pessoa faz o menu,
#       outra faz o registro, outra faz as estatísticas.
# ============================================================

# CÓDIGO DO GRUPO AQUI:

registros = []

def registrar_producao(nome, turno, produzida, defeito):
    taxa_defeito = defeito / produzida if produzida > 0 else 0
    registro = {
        "nome": nome,
        "turno": turno,
        "produzida": produzida,
        "defeito": defeito,
        "taxa_defeito": taxa_defeito
    }
    registros.append(registro)
    print(f"✅ Registro de {nome} adicionado!")

def ver_registros():
    if not registros:
        print("Nenhum registro encontrado.")
        return
    for reg in registros:
        print(f"Operador: {reg['nome']}, Turno: {reg['turno']}, "
              f"Produzida: {reg['produzida']}, Defeito: {reg['defeito']}, "
              f"Taxa de Defeito: {reg['taxa_defeito']:.2f}")

def ver_estatisticas():
    if not registros:
        print("Nenhum registro encontrado.")
        return
    total_produzida = sum(reg['produzida'] for reg in registros)
    total_defeito = sum(reg['defeito'] for reg in registros)
    taxa_media = sum(reg['taxa_defeito'] for reg in registros) / len(registros)
    print(f"Total produzido: {total_produzida}")
    print(f"Total com defeito: {total_defeito}")
    print(f"Taxa média de defeito: {taxa_media:.2f}")

# TESTANDO COM DADOS PRÉ-DEFINIDOS
registrar_producao("Ana Silva", "Manhã", 520, 8)
registrar_producao("Bruno Costa", "Tarde", 495, 15)
registrar_producao("Carla Souza", "Noite", 485, 5)

print("\n=== VER REGISTROS ===")
ver_registros()

print("\n=== ESTATÍSTICAS ===")
ver_estatisticas()


        
````

### 🔶 Atividade em Grupo 2 — Dashboard de Qualidade
**Grupo:** 3-4 pessoas | **Dificuldade:** ⭐⭐⭐

````python
# ============================================================
# 👥 ATIVIDADE EM GRUPO 2 — Dashboard de Qualidade
# ============================================================
#
# ENUNCIADO:
# Com os dados abaixo de uma semana de produção,
# crie um relatório completo que mostre:
#
# 1. Resumo por turno (média de produção, total de defeitos)
# 2. Melhor e pior dia da semana (por taxa de defeito)
# 3. Ranking dos operadores (do mais eficiente ao menos)
# 4. Alerta para dias com taxa de defeito > 3%
# 5. Um "mini-gráfico" de barras com # (ex: Seg: ######## 800)
#
# CRITÉRIO DE AVALIAÇÃO:
# - Código funcional e organizado em funções
# - Output formatado e fácil de ler
# - Comentários explicando o código
# ============================================================

dados_semana = [
    {"dia": "Segunda",  "turno": "Manhã",  "operador": "Ana",    "pecas": 520, "defeitos": 8},
    {"dia": "Segunda",  "turno": "Tarde",  "operador": "Bruno",  "pecas": 495, "defeitos": 18},
    {"dia": "Segunda",  "turno": "Noite",  "operador": "Carla",  "pecas": 485, "defeitos": 5},
    {"dia": "Terça",    "turno": "Manhã",  "operador": "Daniel", "pecas": 510, "defeitos": 12},
    {"dia": "Terça",    "turno": "Tarde",  "operador": "Ana",    "pecas": 530, "defeitos": 6},
    {"dia": "Terça",    "turno": "Noite",  "operador": "Bruno",  "pecas": 340, "defeitos": 25},
    {"dia": "Quarta",   "turno": "Manhã",  "operador": "Carla",  "pecas": 560, "defeitos": 4},
    {"dia": "Quarta",   "turno": "Tarde",  "operador": "Daniel", "pecas": 475, "defeitos": 30},
    {"dia": "Quarta",   "turno": "Noite",  "operador": "Ana",    "pecas": 500, "defeitos": 7},
    {"dia": "Quinta",   "turno": "Manhã",  "operador": "Bruno",  "pecas": 545, "defeitos": 9},
    {"dia": "Quinta",   "turno": "Tarde",  "operador": "Carla",  "pecas": 490, "defeitos": 11},
    {"dia": "Quinta",   "turno": "Noite",  "operador": "Daniel", "pecas": 465, "defeitos": 6},
    {"dia": "Sexta",    "turno": "Manhã",  "operador": "Ana",    "pecas": 580, "defeitos": 3},
    {"dia": "Sexta",    "turno": "Tarde",  "operador": "Bruno",  "pecas": 500, "defeitos": 14},
    {"dia": "Sexta",    "turno": "Noite",  "operador": "Carla",  "pecas": 455, "defeitos": 8},
]

# ============ FUNÇÕES AUXILIARES ============

def calcular_taxa_defeito(pecas, defeitos):
    """Calcula a taxa de defeito em porcentagem."""
    return (defeitos / pecas * 100) if pecas > 0 else 0

def resumo_por_turno(dados):
    """Resumo por turno: média de produção e total de defeitos."""
    turnos = {}
    
    for reg in dados:
        turno = reg["turno"]
        if turno not in turnos:
            turnos[turno] = {"pecas": 0, "defeitos": 0, "registros": 0}
        
        turnos[turno]["pecas"] += reg["pecas"]
        turnos[turno]["defeitos"] += reg["defeitos"]
        turnos[turno]["registros"] += 1
    
    print("\n" + "="*60)
    print("1️⃣  RESUMO POR TURNO")
    print("="*60)
    
    for turno in ["Manhã", "Tarde", "Noite"]:
        if turno in turnos:
            info = turnos[turno]
            media_pecas = info["pecas"] / info["registros"]
            taxa = calcular_taxa_defeito(info["pecas"], info["defeitos"])
            print(f"\n🔷 {turno:8} | Peças: {info['pecas']:4} | "
                  f"Média: {media_pecas:6.1f} | Defeitos: {info['defeitos']:3} | "
                  f"Taxa: {taxa:5.2f}%")

def melhor_pior_dia(dados):
    """Identifica o melhor e pior dia da semana por taxa de defeito."""
    dias = {}
    
    for reg in dados:
        dia = reg["dia"]
        if dia not in dias:
            dias[dia] = {"pecas": 0, "defeitos": 0}
        
        dias[dia]["pecas"] += reg["pecas"]
        dias[dia]["defeitos"] += reg["defeitos"]
    
    # Calcular taxa de defeito por dia
    taxa_dias = {}
    for dia, info in dias.items():
        taxa_dias[dia] = calcular_taxa_defeito(info["pecas"], info["defeitos"])
    
    melhor = min(taxa_dias, key=taxa_dias.get)
    pior = max(taxa_dias, key=taxa_dias.get)
    
    print("\n" + "="*60)
    print("2️⃣  MELHOR E PIOR DIA DA SEMANA")
    print("="*60)
    print(f"✅ MELHOR: {melhor:8} | Taxa: {taxa_dias[melhor]:6.2f}%")
    print(f"❌ PIOR:   {pior:8} | Taxa: {taxa_dias[pior]:6.2f}%")

def ranking_operadores(dados):
    """Ranking dos operadores do mais eficiente ao menos."""
    operadores = {}
    
    for reg in dados:
        op = reg["operador"]
        if op not in operadores:
            operadores[op] = {"pecas": 0, "defeitos": 0}
        
        operadores[op]["pecas"] += reg["pecas"]
        operadores[op]["defeitos"] += reg["defeitos"]
    
    # Calcular taxa para cada operador
    ranking = []
    for op, info in operadores.items():
        taxa = calcular_taxa_defeito(info["pecas"], info["defeitos"])
        ranking.append((op, taxa, info["pecas"], info["defeitos"]))
    
    # Ordenar por taxa de defeito (crescente = mais eficiente)
    ranking.sort(key=lambda x: x[1])
    
    print("\n" + "="*60)
    print("3️⃣  RANKING DOS OPERADORES")
    print("="*60)
    
    for i, (op, taxa, pecas, defeitos) in enumerate(ranking, 1):
        emoji = "🥇" if i == 1 else "🥈" if i == 2 else "🥉" if i == 3 else f"{i}️⃣ "
        status = "✅ Excelente" if taxa <= 2 else "⚠️  Atenção" if taxa <= 4 else "🔴 Crítico"
        print(f"{emoji} {op:8} | Taxa: {taxa:6.2f}% | Peças: {pecas:4} | "
              f"Defeitos: {defeitos:3} | {status}")

def alertas_taxa_defeito(dados):
    """Alertas para dias com taxa de defeito > 3%."""
    dias = {}
    
    for reg in dados:
        dia = reg["dia"]
        if dia not in dias:
            dias[dia] = {"pecas": 0, "defeitos": 0}
        
        dias[dia]["pecas"] += reg["pecas"]
        dias[dia]["defeitos"] += reg["defeitos"]
    
    print("\n" + "="*60)
    print("4️⃣  ALERTAS — TAXA DE DEFEITO > 3%")
    print("="*60)
    
    alertas = []
    for dia, info in dias.items():
        taxa = calcular_taxa_defeito(info["pecas"], info["defeitos"])
        if taxa > 3.0:
            alertas.append((dia, taxa, info["defeitos"]))
    
    if alertas:
        alertas.sort(key=lambda x: x[1], reverse=True)
        for dia, taxa, defeitos in alertas:
            print(f"🚨 {dia:8} | Taxa: {taxa:6.2f}% | Defeitos: {defeitos:3}")
    else:
        print("✅ Nenhum dia com taxa superior a 3%!")

def mini_grafico_barras(dados):
    """Mini-gráfico de barras com # representando a produção."""
    dias_ord = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"]
    dias = {}
    
    for reg in dados:
        dia = reg["dia"]
        if dia not in dias:
            dias[dia] = 0
        dias[dia] += reg["pecas"]
    
    print("\n" + "="*60)
    print("5️⃣  MINI-GRÁFICO DE PRODUÇÃO")
    print("="*60)
    
    for dia in dias_ord:
        if dia in dias:
            pecas = dias[dia]
            # Escala: cada # representa 50 peças
            barras = "#" * (pecas // 50)
            print(f"{dia:8} | {barras} {pecas}")

# ============ EXECUÇÃO PRINCIPAL ============
print("\n" + "🏭 "*30)
print("DASHBOARD DE QUALIDADE — SEMANA DE PRODUÇÃO")
print("🏭 "*30)

resumo_por_turno(dados_semana)
melhor_pior_dia(dados_semana)
ranking_operadores(dados_semana)
alertas_taxa_defeito(dados_semana)
mini_grafico_barras(dados_semana)

print("\n" + "="*60)
print("✅ Relatório Finalizado!")
print("="*60)
````
