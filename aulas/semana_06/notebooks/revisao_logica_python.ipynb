## 🔷 Bloco 3 — Operadores

### Operadores Aritméticos (contas)

| Operador | Significado | Exemplo |
|----------|-------------|---------|
| `+` | Soma | `10 + 5 = 15` |
| `-` | Subtração | `10 - 3 = 7` |
| `*` | Multiplicação | `4 * 3 = 12` |
| `/` | Divisão | `10 / 4 = 2.5` |
| `//` | Divisão inteira | `10 // 4 = 2` |
| `%` | Resto da divisão | `10 % 3 = 1` |
| `**` | Potência | `2 ** 8 = 256` |

````python
# =============================================
# 🏭 CENÁRIO: Calculando indicadores de produção
# =============================================

pecas_produzidas = 1500
pecas_com_defeito = 37
horas_trabalhadas = 8
numero_operadores = 6
custo_por_peca = 12.50  # reais

# Peças produzidas
print(f"Peças produzidas: {pecas_produzidas}")

# Peças boas
pecas_boas = pecas_produzidas - pecas_com_defeito
print(f"Peças boas: {pecas_boas}")

# Taxa de defeito (em %)
taxa_defeito = (pecas_com_defeito / pecas_produzidas) * 100
print(f"Taxa de defeito: {taxa_defeito:.2f}%")

# Produtividade por operador
produtividade = pecas_boas / numero_operadores
print(f"Peças boas por operador: {produtividade:.1f}")

# Produção por hora
pecas_por_hora = pecas_boas / horas_trabalhadas
print(f"Peças boas por hora: {pecas_por_hora:.1f}")

# Faturamento do turno
faturamento = pecas_boas * custo_por_peca
print(f"Faturamento do turno: R$ {faturamento:,.2f}")
````

````python
# Operadores de comparação — retornam True ou False
print("=== COMPARAÇÕES ===")
print(f"1500 > 1000? {1500 > 1000}")    # Maior que
print(f"37 < 50? {37 < 50}")            # Menor que
print(f"taxa_defeito <= 5? {taxa_defeito <= 5.0}")  # Menor ou igual
print(f"turno == 'Tarde'? {'Manhã' == 'Tarde'}")   # Igual
print(f"turno != 'Tarde'? {'Manhã' != 'Tarde'}")   # Diferente
````

````python
# =============================================
# 🏭 CENÁRIO: Sistema de alertas de qualidade
# =============================================

taxa_defeito = 3.2  # tente mudar esse valor!

print(f"Taxa de defeito atual: {taxa_defeito}%")
print()

if taxa_defeito <= 2.0:
    status = "✅ EXCELENTE"
    acao = "Manter padrão atual"
elif taxa_defeito <= 5.0:
    status = "⚠️  ATENÇÃO"
    acao = "Monitorar de perto e revisar processo"
elif taxa_defeito <= 10.0:
    status = "🔴 CRÍTICO"
    acao = "Acionar supervisor imediatamente"
else:
    status = "🚨 PARAR LINHA"
    acao = "Interromper produção e chamar engenharia"

print(f"Status: {status}")
print(f"Ação recomendada: {acao}")
````

````python
# Usando operadores lógicos: and, or, not

temperatura = 120.0
pressao = 2.2
valvula_aberta = False

# and = as DUAS condições precisam ser verdadeiras
if temperatura > 180 and pressao > 3.0:
    print("⚙️  Processo dentro do padrão")

# or = PELO MENOS UMA condição precisa ser verdadeira
if temperatura > 220 or pressao > 5.0:
    print("🚨 Risco de superaquecimento!")
else:
    print("✅ Temperatura e pressão normais")

# not = inverte o valor lógico
if not valvula_aberta:
    print("❌ Abra a válvula antes de iniciar!")
else:
    print("✅ Válvula aberta — pode iniciar")
````

## 🔷 Bloco 6 — Estruturas de Repetição

### 6.1 — Loop `for`
Usamos `for` quando sabemos **quantas vezes** queremos repetir algo.

````python
# Usando range() — gera sequências de números
print("Contagem de 1 a 5:")
for numero in range(1, 6):
    print(numero)

print("\nTabuada do 3:")
for i in range(0, 11):
    resultado = 3 * i
    print(f"3 x {i:2} = {resultado:3}")
````

````python
# for com if dentro — combinando tudo!
producao_semanal = [1450, 1380, 1520, 1490, 1600]
dias = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"]
meta_diaria = 1500

print("=== ANÁLISE DA META ===")
dias_acima = 0
dias_abaixo = 0

for i in range(len(dias)):
    producao = producao_semanal[i]
    if producao >= meta_diaria:
        print(f"✅ {dias[i]}: {producao} — META BATIDA")
        dias_acima += 1
    else:
        diferenca = meta_diaria - producao
        print(f"❌ {dias[i]}: {producao} — faltaram {diferenca} peças")
        dias_abaixo += 1

print(f"\nDias acima da meta: {dias_acima}")
print(f"Dias abaixo da meta: {dias_abaixo}")
````

### 6.2 — Loop `while`
Usamos `while` quando queremos repetir **enquanto uma condição for verdadeira**.

> ⚠️ **Cuidado com loops infinitos!** A condição precisa se tornar falsa em algum momento.

````python
# =============================================
# 🏭 CENÁRIO: Simulando uma esteira de produção
# =============================================

estoque_materias_primas = 100  # kg de matéria-prima
consumo_por_lote = 15          # kg consumidos por lote
lotes_produzidos = 0

print("=== SIMULAÇÃO DE PRODUÇÃO ===")
print(f"Estoque inicial: {estoque_materias_primas} kg\n")

while estoque_materias_primas >= consumo_por_lote:
    estoque_materias_primas -= consumo_por_lote
    lotes_produzidos += 1
    print(f"Lote {lotes_produzidos} produzido | Estoque restante: {estoque_materias_primas} kg")

print(f"\n⚠️  Matéria-prima insuficiente para novo lote")
print(f"Total de lotes produzidos: {lotes_produzidos}")
print(f"Sobra de material: {estoque_materias_primas} kg")
````

## 🔷 Bloco 8 — Funções

Funções são **blocos de código reutilizáveis**. Em vez de repetir o mesmo código, você cria uma função e chama ela quando precisar.

```python
def nome_da_funcao(parametro1, parametro2):
    # código aqui
    return resultado
```

````python
# =============================================
# 🏭 CENÁRIO: Funções para indicadores de qualidade
# =============================================

def calcular_taxa_defeito(total_produzido, total_defeituoso):
    """Calcula a taxa de defeito em porcentagem."""
    if total_produzido == 0:
        return 0
    taxa = (total_defeituoso / total_produzido) * 100
    return taxa


def classificar_qualidade(taxa_defeito):
    """Classifica a qualidade com base na taxa de defeito."""
    if taxa_defeito <= 1.0:
        return "EXCELENTE"
    elif taxa_defeito <= 3.0:
        return "BOM"
    elif taxa_defeito <= 5.0:
        return "ATENÇÃO"
    else:
        return "CRÍTICO"


def gerar_relatorio_turno(turno, operador, pecas, defeitos):
    """Gera um relatório completo do turno."""
    taxa = calcular_taxa_defeito(pecas, defeitos)
    classificacao = classificar_qualidade(taxa)
    pecas_boas = pecas - defeitos
    
    print(f"\n{'='*40}")
    print(f"  RELATÓRIO — TURNO {turno.upper()}")
    print(f"{'='*40}")
    print(f"  Operador      : {operador}")
    print(f"  Peças prod.   : {pecas}")
    print(f"  Peças boas    : {pecas_boas}")
    print(f"  Defeituosas   : {defeitos}")
    print(f"  Taxa defeito  : {taxa:.2f}%")
    print(f"  Classificação : {classificacao}")
    print(f"{'='*40}")


# Chamando as funções
gerar_relatorio_turno("Manhã", "Ana Silva", 520, 8)
gerar_relatorio_turno("Tarde", "Bruno Costa", 495, 28)
gerar_relatorio_turno("Noite", "Carla Souza", 485, 4)
````

## 📚 Resumão — O que aprendemos hoje?

| Conceito | O que é | Quando usar |
|----------|---------|-------------|
| **Variáveis** | Guardam valores | Sempre que precisar armazenar dados |
| **Tipos de dados** | int, float, str, bool | Depende do tipo de informação |
| **if/elif/else** | Tomada de decisão | Quando o código deve seguir caminhos diferentes |
| **for** | Repetir N vezes | Quando você sabe quantas iterações fazer |
| **while** | Repetir enquanto... | Quando a parada depende de uma condição |
| **Listas** | Coleção de valores | Vários itens do mesmo tipo |
| **Dicionários** | Chave: valor | Representar objetos/registros |
| **Funções** | Código reutilizável | Quando você repete a mesma lógica |
